```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cm_01 = pd.read_csv('./cmatrix_01.csv', header=None)
cm_02 = pd.read_csv('./cmatrix_02.csv', header=None)
```

```python
def plot_cmatrix(cm):
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    print(cm)
    print(cm_normalized)

    f, ax=plt.subplots(figsize=(7,7))
    sns.heatmap(cm,annot=True,linewidths=0.5,linecolor="red",ax=ax,
                fmt='.2f', xticklabels=['0', '1', '2', '3', '4', '5'], yticklabels=['0', '1', '2', '3', '4', '5'], cmap="YlGnBu")
    plt.title("Confusion Matrix")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show(block=False)

    f, ax=plt.subplots(figsize=(7,7))
    sns.heatmap(cm_normalized,annot=True,linewidths=0.5,linecolor="red",ax=ax,
                fmt='.2f', xticklabels=['0', '1', '2', '3', '4', '5'], yticklabels=['0', '1', '2', '3', '4', '5'], cmap="YlGnBu")
    plt.title("Normalized Confusion Matrix")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show(block=False)
```

```python
plot_cmatrix(cm_01)
```

```python
plot_cmatrix(cm_02)
```
